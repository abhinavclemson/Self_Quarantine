```python
from plotly import tools
import plotly.offline as py
import plotly.graph_objs as go


trace1 = go.Pie(
     values=[16, 15, 12, 6, 5, 4, 42],
     labels=["US", "China", "European Union", "Russian Federation",
             "Brazil", "India", "Rest of World"
             ],
     domain=dict(x=[0.30, 0.55]),
     name="GHG Emissions",
     hoverinfo="label+percent+name",
)
trace2 = go.Pie(
     values=[27, 11, 25, 8, 1, 3, 25],
     labels=["US", "China", "European Union", "Russian Federation","Brazil", "India", "Rest of World"],
     domain=dict(x=[0, 0.25]),
     name="CO2 Emissions",
     hoverinfo="label+percent+name",
)

trace3= go.Histogram(
            y=[16, 15, 12, 6, 5, 4, 42],
            x=["US", "China", "European Union", "Russian Federation",
             "Brazil", "India", "Rest of World"],
            domain=dict(x=[0, 0.25]),

)

layout = go.Layout(title="Global Emissions 1990-2011",)
data = [trace1, trace2,trace3]
fig = go.Figure(data=data, layout=layout)
py.plot(fig)
```
